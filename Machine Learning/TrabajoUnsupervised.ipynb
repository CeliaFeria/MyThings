# Hierarchical clustering

```python
#importar librerias 
import scipy.cluster.hierarchy as shc
from matplotlib import pyplot as plt

#fijamos el tamaño de la figura y el título
plt.figure(figsize=(20, 17))  
plt.title("Single Dendogram")  

#eliminamos la cuadrícula para visualizar mejor 
ax = plt.axes()        
ax.xaxis.grid(False)
ax.yaxis.grid(False)

#llevamos a cabo el hierarchichal clustering con el método single y la medida de distancia euclidean
#visualizamos con shc.dendrogram
dend = shc.dendrogram(shc.linkage(Data_t.values, method='single', metric='euclidean'), labels=genes)
  
```

```python
plt.figure(figsize=(20, 17))  
plt.title("Ward Dendogram") 

ax = plt.axes()        
ax.xaxis.grid(False)
ax.yaxis.grid(False)

#llevamos a cabo el hierarchichal clustering con el método ward y la medida de distancia euclidean
#visualizamos con shc.dendrogram
dend = shc.dendrogram(shc.linkage(Data_t.values, method='ward', metric='euclidean'), labels=genes)  
```

```python
plt.figure(figsize=(20, 17))  
plt.title("Complete Dendrogram") 

ax = plt.axes()        
ax.xaxis.grid(False)
ax.yaxis.grid(False)


#llevamos a cabo el hierarchichal clustering con el método complete y la medida de distancia euclidean
#visualizamos con shc.dendrogram
dend = shc.dendrogram(shc.linkage(Data_t.values, method='complete', metric='euclidean'), labels=genes)
```

```python
plt.figure(figsize=(20, 17))  
plt.title("Complete Dendrogram") 

ax = plt.axes()        
ax.xaxis.grid(False)
ax.yaxis.grid(False)

#llevamos a cabo el hierarchichal clustering con el método complete y la medida de distancia chebyshev
#visualizamos con shc.dendrogram
dend = shc.dendrogram(shc.linkage(Data_t.values, method='complete', metric='chebyshev'), labels=genes)
```

# Partitional clustering

```python
#importamos librerias
from sklearn.cluster import KMeans

#generamos la agrupacion de los datos el método k-means y un valor de k igual a 3
model = KMeans(n_clusters=3)
model.fit(Data_t)
#guardamos la predicción
label= model.predict(Data_t)

Data2 = Data_t

#añadimos una columna con la predicción de cluster para cada gen
Data2['Cluster'] = label

#visualización de la tabla
print(Data2.head())
```

```python
#valores de inertia
model.inertia_
print(label)
```

```python
from sklearn.decomposition import PCA

# Utilizamos el método de PCA para 
reduced_data = PCA(n_components = 2).fit_transform(Data_t)
#guardamos los resultados de los componentes generados
results = pd.DataFrame(reduced_data,columns=['pca1','pca2'])

x = results['pca2']
y = results['pca1']

#visualización de los componentes según el cluster, diferenciados por colores
sns.scatterplot(x=x, y=y, hue= label)
plt.title('K-means Clustering with 2 dimensions')
plt.show()
```

```python
#generamos la agrupacion de los datos el método k-means y un valor de k igual a 5
model = KMeans(n_clusters=5) 
model.fit(Data2)
#guardamos las predicciones
label= model.predict(Data2)

#añadimos a la tabla estas predicciones
Data2['Cluster'] = label

print(Data_t.head())
```

```python
model.inertia_
```

```python
from sklearn.decomposition import PCA

# Utilizamos el método de PCA para 
reduced_data = PCA(n_components = 2).fit_transform(Data_t)
#guardamos los resultados de los componentes generados
results = pd.DataFrame(reduced_data,columns=['pca1','pca2'])

x = results['pca2']
y = results['pca1']

#visualización de los componentes según el cluster, diferenciados por colores
sns.scatterplot(x=x, y=y, hue= label)
plt.title('K-means Clustering with 2 dimensions')
plt.show()
```

Elbow Plot

```python
#creamos un array para guardar los distintos valores de inertias según el valor de k 
inertias = []
#rango de valores de k
K = range(1,11)
# iteramos para obtener los distinos valores de inertia
for k in K:
    model = KMeans(n_clusters=k) 
    model.fit(Data_t)
    inertias.append(model.inertia_)
```

```python
#asignamos tamaño a la figura
plt.figure(figsize=(16,8))
plt.plot(K, inertias, 'x-')
#añadimos títulos 
plt.xlabel('k')
plt.ylabel('Inertias')
plt.title('Optimal k')
#visualización
plt.show()
```

```python
#Make clustering with the optimal value K=4

model = KMeans(n_clusters=4) 
model.fit(Data_t)
label= model.predict(Data_t)

Data_t['Cluster'] = label

print(Data_t.head())
```

```python
from sklearn.decomposition import PCA

# Utilizamos el método de PCA para 
reduced_data = PCA(n_components = 2).fit_transform(Data_t)
#guardamos los resultados de los componentes generados
results = pd.DataFrame(reduced_data,columns=['pca1','pca2'])

x = results['pca2']
y = results['pca1']

#visualización de los componentes según el cluster, diferenciados por colores
sns.scatterplot(x=x, y=y, hue= label)
plt.title('K-means Clustering with 2 dimensions')
plt.show()
```

Silhouette score

```python
#importamos librerias
from sklearn.metrics import silhouette_score
#creamos array para guardar los distintos valores de silhoutte
sil = []
#rango de valores de k
K = range(2,11)

#iteramos para obtener los distintos valores de silhoutte según el valore de k
for k in K:
    model = KMeans(n_clusters=k) 
    model.fit(Data_t)
    label= model.predict(Data_t)
    score = silhouette_score(Data_t, label)
    sil.append(score)
```

```python
#visualización de los distintos valores de silhoutte
plt.figure(figsize=(16,8))
plt.plot(K, sil, 'bx-')
plt.xlabel('k')
plt.ylabel('Silhouette')
plt.title('Optimal k')
plt.show()
```

```python
#importar librerias 
from yellowbrick.cluster import silhouette_visualizer

#visualización de los valores de silhoutte para un determinado número de clusters
silhouette_visualizer(KMeans(3), Data_t, colors='yellowbrick')

silhouette_visualizer(KMeans(4), Data_t, colors='yellowbrick')

silhouette_visualizer(KMeans(5), Data_t, colors='yellowbrick')
```

Relacion con tejido

```python
#importar librerias
from sklearn.cluster import KMeans
from collections import Counter

#seleccionamos 9 cluster, ya que son el número de tejidos que hay en los datos
kmeans = KMeans(n_clusters=9) 
kmeans = kmeans.fit(Data)
labels = kmeans.predict(Data)
centroids  = kmeans.cluster_centers_
Counter(labels)

```

```python
#redondeamos los valores de los centroides 
round_to_tenths = [round(num, 1) for num in centroids[0]]
print(round_to_tenths)
```

```python
# obtenemos los features más importantes para la división
idx = 0
for value in centroids[0]:
    round_value = round(value,1)
    if (round_value < 0.0):
        feature = Data.columns[idx]
        print(feature,value)
    idx+=1
```
